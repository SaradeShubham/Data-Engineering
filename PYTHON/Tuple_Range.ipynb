# tuple

- Tuple are immutable ( un Chandable),ordered sequence of items

- They are defined using () , item is sepersted by comas.

- like list they can contains item diff type.

- Because  they are immutable they are use to fixed collection of items.

- Tuple support indexing , slicing,len() and membership testing(in, not in) like list.

- They lack methods that modifies the tuple like remove,extend,append ect.

- They have only two primary methods which is count() and index.

#### LOOP
- What is a for loop?
- A for loop is used when you want to reapeat some code for every iteme in a sequence(list,tuple,string or range)

for iteme in sequense :
   -do somethin

- for--> start the loop
- item--> tempory variable
- in--> used to access each element inside the sequence
- sequen--> can be the list , tuple whatever data is it


```python
print("HELLO WORLD")
print("HELLO WORLD")
print("HELLO WORLD")
print("HELLO WORLD")
print("HELLO WORLD")
print("HELLO WORLD")
print("HELLO WORLD")
```

```python
for i in range(50):
    print("Hello World")
```

```python
fruits=["Apple","Banana","cherry"] ## sequence

for fruit in fruits:
    print(fruit)
```

```python
for i in fruits:
    print(i)
```

```python
for letter in "Shubham":
    print(letter)
```

```python
name=["lohit","nandini","shubham"]
upper_name=[]

for i in name:
    upper_name.append(i.upper())

print(name)
print(upper_name)
```

```python
mixed= [10,"hello",25,"world",40,"AI"]

number=[]
word=[]

for item in mixed:
    if type(item)==int:
        number.append(item)
    else:
        word.append(item)

print(number)
print(word)
```

##Range
- Represent an immutable sequence of number, typically used for looping a specific number of time for loop.

- Its memory efficient because it doesnt generate all the numbers at once, only when needed(its generator).

- range(stop), range(start,stop) or range(start,stop,step)

- ##### Commom Operation on range

-  len(),membership testing(in,not in),indexing,slicing.

```python
## creation and itteration

# creating range objects and using them in for loops.

# simple range (0 up to,but not including,5)
#range(stop)==
for i in range(5):
    print(f"- iteration{i}")
```

```python
#range with start and stop
## create batch id from 101 to 105

batch_ids= range(101,106)
for batch_id in batch_ids:
    print(f"-proccesing batch: {batch_id}")
```

```python
# Range (start,stop,step)
## for finding even numbers

numbers=range(0,11,2)

for even_number in numbers:
    print(f"-Even Number:{even_number}")
```

```python
## For odd numbers

numbers=range(1,11,2)

for odd_number in numbers:
    print(f"-Even Number:{odd_number}")
```

```python
## count down

countdown=range(5,0,-1)

for i in countdown:
    print(i)
```

```python
course_list=["DS","DE","ADO"]

for i in range(len(course_list)):
    print(f"-index{i}:{course_list[i]}")
```
