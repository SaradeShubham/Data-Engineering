### Dictionary

##### Mapping Type(dict) {key:value}

- Dictonary (dict) store collections of key:value pairs.

- Key must be unique and immutable(like string, numbet, tuple etc.)

- Values can be of any type and are accessed using their corresponding keys.

- Dictionaries are mutable.

## pop()

- If the key is not found , it raises a key error unless a default value is not provided.

- If defalt value is provided and key is not found , it returns the default value.

- Modifies the dictionary in place.
