- Modifies the liat in-place.

```python
course_list=["Python Baasics","Data Science","Data Engineering"]

tasks=["Review Code","Update Task","Deploy"]
```

#sort Function

- Syntaxt= list.sort(key=none,  reverse =True or False)

- Sort the item of list in-place.

- Key= A function to be called on each element before macking copaarison(key= str.lower() for case-sensitive sort.)

- reverse -= True , Sort in Descending order.

```python
course_list=["Data Engineering","Azure DevOps","Data Science","Python Basics"]

student_count=[350,420,380,500,450]

mixed_case_course= ["Azure DevOps","Python Basics","Data SCience"]
```

```python
## sort list by numerical order
print(student_count)
student_count.sort(reverse=True)

print(student_count)
```

```python
mixed_case_course= ["Azure DevOps","Python Basics","Data SCience","basics python"]
print(mixed_case_course)

mixed_case_course.sort(key=str.lower, reverse=True)

print(mixed_case_course)

```
