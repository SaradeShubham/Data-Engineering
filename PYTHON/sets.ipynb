##Set types

- These are unordered coolection of unique items.
- They are useful for membership testing, removing duplicates, and performing mathematical oprerations.
  (like union , intersections, difference.)

- Created using set()
- Sets are mutable.

```python
list=[]
tuple=()
set={}

print(type(list))
print(type(tuple))
print(type(set))
```

```python
set1=set()

print(type(set1))
```

```python
set1={"name","Shubham"}

print(type(set1))
```

### Update(iterable)

- Adds all unique ites from an iterable (like list,tuple,set or string ) into the set ans its modifies the set

```python
devops_tools={"Docker","Kubernetes"}
data_tools={"Python","SQL","Pandas"}
new_tools_list=["Terraform","GitHub Aactions","Docker"] ## here a duplicate "Docker" is present
more_skills=("SQL","Git") ## here "SQL" duplicate is there
```

```python
## Update devops_tool eith new_tools_list
devops_tools.update(new_tools_list)
devops_tools
```

```python
data_tools.update(more_skills)
data_tools
```

```python
## Update with characters from a string

char={"a","b"}
char.update("ace")
char
```

```python
## update with an empty data type
initial_set={1,2}
initial_set.update([])
initial_set
```

```python
empty_set=set()
empty_set.update([101,102,101])
empty_set
```

```python
## update with items from a range

numbers= {1,5}
numbers.update(range(3,7))
print(numbers)
```

```python
## update with multiple iterable at once

features={"Login"}
features.update(["Profile", "Settings"],("Logout","Profile"))
features
```

## remove(items)
- Removes item.Rises a KeyErro if item is not found

## discard(items)
- Remove items if present.Does not rise an error if item is not found.

- Both remove and discard modify the set in place.

```python
devops_tools={"Docker","Kubernetes","Terraform","GitHub Actions"}
```

```python
devops_tools.remove("Terraform")
devops_tools
```

```python
devops_tools.remove("Terraform")## here item is not present in the set therefor it throws an rerror
devops_tools
```

```python
devops_tools={"Docker","Kubernetes","Terraform","GitHub Actions"}
devops_tools.discard("Kubernetes")
devops_tools
```

```python
devops_tools={"Docker","Terraform","GitHub Actions"}
devops_tools.discard("Kubernetes")## here "KUbernetes" is not present but it does not throw any type of error
devops_tools
```

### pop()

- REmoves and return an element form the set.add.

- Rises an error if set is empty.

- Since set is unordered you can not predict which element gets pop.

- Modifies the set in-place.

```python
data_analytics_topics={"Visualization","Power Bi","EDA","Reporting"}

numbers={101,102,103}
```

```python
poped_topic=data_analytics_topics.pop()
poped_topic
```

```python
poped_topic=data_analytics_topics.pop()
poped_topic
```

```python
poped_topic=data_analytics_topics.pop()
poped_topic
```

```python
data_analytics_topics  ## it modifies set in place
```

```python
pop_numbers=numbers.pop()## if set got empty it throws error
pop_numbers
```
