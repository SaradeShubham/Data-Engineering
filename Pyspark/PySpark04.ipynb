```python
## to acces reide count of eache bike type(rideable type) having
from pyspark.sql.functions import * ## all functions are imoported like um,count,min,max all like thar it imports
df1=citibike_df.groupBy("rideable_type").agg(count("rideable_type").alias("Total_Rides"))

df1.show()
```

```python
## to find out min ride id and the max ride id as per the rideable type
df1= citibike_df\
          .groupBy("rideable_type")\
          .agg(count("rideable_type").alias("total_rides"),
               min("ride_id").alias("min_ride_id"),
               max("ride_id").alias("max_ride_id"))

df1.show()

```

```python
df1= citibike_df\
          .groupBy("rideable_type")\
          .agg(count("rideable_type").alias("total_rides"),
               min("ride_id").alias("min_ride_id"),
               max("ride_id").alias("max_ride_id"))\
          .orderBy("rideable_type") #asccending order

df1.show()
```

```python
df1= citibike_df\
          .groupBy("rideable_type")\
          .agg(count("rideable_type").alias("total_rides"),
               min("ride_id").alias("min_ride_id"),
               max("ride_id").alias("max_ride_id"))\
          .orderBy(desc("rideable_type"))## Descending order(in spark we have to mention column name to aplly desc() function)

df1.show()

df1.printSchema()
```

```python
## in schema of total rides we got data type as long but i want convert int into the integer data type .
from pyspark.sql.functions import *, col
from pyspark.sql.types import IntegerType
## withCollumn and Casting the column
df1= citibike_df\
          .groupBy("rideable_type")\
          .agg(count("rideable_type").alias("total_rides"),
               min("ride_id").alias("min_ride_id"),
               max("ride_id").alias("max_ride_id"))\
          .orderBy(desc("rideable_type"))

df1.show()

df1.printSchema()

df2=df1.withColumn("total_rides", col("total_rides").cast(IntegerType()))

df2.printSchema()
```

```python

##with column wi able to make new column.S
df2=df1.withColumn("total_rides_new", col("total_rides").cast(IntegerType()))
df2.show()
df2.printSchema()
```

```python
## Renaming the column

### withColumnRenamed fuction we reamne the perticular column name .
df2=df1.withColumn("total_rides", col("total_rides").cast(IntegerType()))\
       .withColumnRenamed("rideable_type", "Bike_type")
df2.show()
df2.printSchema()
```

```python
## there are 4 wite mode in the spark

#1. overwrite mode==> Replaces existing data with new data.

#2. append mode==>  Adds new data to the existing dataset without removing old data.

#3. ignore mode==>  If data already exists at the destination, Spark does nothing (ignores the write).

#4. Eroor mode==>  Throws an error if data already exists at the destination.
```

```python
## Overwite mode==>sytax==> mode(" mode")
citibike_df.createOrReplaceTempView("citibike")

result=spark.sql("select rideable_type, \
                        count(*) as total_ride, \
                        min (ride_id) min_ride, \
                        max(ride_id) max_ride \
                        from citibike group by rideable_type")
result.show()
## for now i want to save this result.

result.write.mode("overwrite").csv("results", header=True)
```

```python
## append mode==>sytax==> mode(" mode")
citibike_df.createOrReplaceTempView("citibike")

result=spark.sql("select rideable_type, \
                        count(*) as total_ride, \
                        min (ride_id) min_ride, \
                        max(ride_id) max_ride \
                        from citibike group by rideable_type")
result.show()
## for now i want to save this result.

result.write.mode("append").csv("results", header=True)
```

```python
## ignor mode==>sytax==> mode(" mode")
citibike_df.createOrReplaceTempView("citibike")

result=spark.sql("select rideable_type, \
                        count(*) as total_ride, \
                        min (ride_id) min_ride, \
                        max(ride_id) max_ride \
                        from citibike group by rideable_type")
result.show()
## for now i want to save this result.

result.write.mode("ignore").csv("results", header=True)
```
