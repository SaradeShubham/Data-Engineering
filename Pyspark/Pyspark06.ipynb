```python
##Joins

## to perform join function there is atleast one common column is required in both table.
## In single table is also we can perform join function.


## Types of Join==>

# 1.Inner Join==> Returns rows with matching keys in both DataFrames. Default join type.

# 2. Left Outer Join==> Returns all rows from the left DataFrame and matching rows from the right.

# 3. Right Outer Join==> Returns all rows from the right DataFrame and matching rows from the left.

# 4. Full Outer Join==> Returns all rows from both DataFrames, filling unmatched values with null.

# 5. Left Semi Join ==> Returns rows from the left DataFrame that have matches in the right.

# 6. Left Anti Join==> Returns rows from the left DataFrame that do not have matches in the right.

# 7. Cross Join==> Cartesian product of both DataFrames (all combinations).

# 8. Self Join ==> Join a DataFrame with itself.
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("customer_name", StringType(), True),
    StructField("city", StringType(), True)
])


data=[(1,"Alice","Bengaluru"),
(2,"Bob","Mumbai"),
(3,"Charlie","Delhi"),
(4,"Diana","Pune"),
(5,"Ethan","Hyderabad")]

columns=("customer_id","customer_name","city")
df1=spark.createDataFrame(data,schema=schema)

df1.show()

df1.write.mode("overwrite").csv("test_join")

df1.printSchema()
```

```python

schema = StructType([
    StructField("order_id", IntegerType(), True),
    StructField("customer_id", IntegerType(), True),
    StructField("product", StringType(), True),
    StructField("amount", IntegerType(), True)
])

columns=("order_id","customer_id","product","amount")
data=[(101,1,"Laptop",75000),
      (102,2,"Mobile",25000),
      (103,2,"Headphones",2000),
      (104,3,"Tablet",30000),
      (105,6,"Smartwatch",12000)]

df2=spark.createDataFrame(data,schema=schema)

df2.show()

df2.write.mode("append").csv("test_join")



```

```python
df2.show()
df2.printSchema()
```

```python
df1.show()
df2.show()
```

```python
##join==> default as Inner Join

df1.join(df2,df1.customer_id==df2.customer_id).show()
```

```python
## Left Join

df1.join(df2,df1.customer_id==df2.customer_id,"left").show()
```

```python
df2.join(df1,df2.customer_id==df1.customer_id,"left").show()
```

```python
## Right join

df1.join(df2,df1.customer_id==df2.customer_id,"right").show()
```

```python
df2.join(df1,df2.customer_id==df1.customer_id,"right").show() ## here right side table is df1.
```

```python
## Full join
df1.join(df2,df1.customer_id==df2.customer_id,"full").show() ## keeping all data from both tables.
```

```python
df2.join(df1,df2.customer_id==df1.customer_id,"full").show()
```

```python
### now we jus want to find out the common customer id only.
## Left semi join==> give coomon value only fron left table
## it removes the unmatched row and also columns.
df1.join(df2,df1.customer_id==df2.customer_id,"left_semi").show()



```

```python
## Left anti join==> gives only unmatched row and column from our data

df1.join(df2,df1.customer_id==df2.customer_id,"left_anti").show()
```

```python
## join as basis on the column only .
df1.join(df2,df1.customer_id==df2.customer_id).select("customer_name","city").show()
```

```python
df1.join(df2,df1.customer_id==df2.customer_id).select("product","amount").show()
```

```python
### cross join

df1.crossJoin(df2).show()
```

```python
cross_join=df1.crossJoin(df2)

cross_join.show()

print(cross_join.count())
```
